```python
languages = ['Farsi_Persian-UTF8','Esperanto-UTF8',
             'IrishGaelic_Gaeilge-Latin1', 
             'Chinese_Mandarin-GB2312',
             'TokPisin-Latin1']
```

```python
allLangs = [udhr.raw(lang) for lang in languages] 
```

```python
allTokens = [nltk.tokenize.regexp_tokenize(lang, '\w+')
             for lang in allLangs]
```

```python
len(allTokens)
```

```python
lengths = [len(text) for text in allTokens]
```

```python
pd.Series(lengths, index=languages).plot(kind="bar")
```
